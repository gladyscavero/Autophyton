```python
database = '/content/drive/MyDrive/Ciclo 9/AutoPython/Archivos Importantes/Database.csv'


d1 = {
    'prokaryotic': {   
        'prom' :{
            'fullpro':{},
            'minpro':{}
            },

        'RBD':{},
        'term':{}
},
    'eukaryotic(mammals)' : {   
        'prom' :{
            'fullpro':{},
            'minpro':{}
            },

        'RBD':{},
        'term':{}
},
      'eukaryotic(plants)':{   
        'prom' :{
            'fullpro':{},
            'minpro':{}
            },

        'RBD':{},
        'term':{}
}, 'box':{}

}



import csv
with open(database, 'r') as file:
    reader = csv.reader(file)
    
    for row in reader:
      print(row)
      row[4]= row[4].upper()
      if row[1] == 'prokaryotic':
        if row[3] == 'fullpro':
          d1['prokaryotic']['prom']['fullpro'][row[0]] = row[4]
        elif row[3] == 'minpro':
          d1['prokaryotic']['prom']['minpro'][row[0]] = row[4]
        elif row[3] == 'RBD':
          d1['prokaryotic']['RBD'][row[0]] = row[4]
        elif row[3] == 'terminator':
          d1['prokaryotic']['term'][row[0]] = row[4]

      elif row[2] == 'plant':
          if row[3] == 'fullpro':
            d1['eukaryotic(plants)']['prom']['fullpro'][row[0]] = row[4]
          elif row[3] == 'minpro':
            d1['eukaryotic(plants)']['prom']['minpro'][row[0]] = row[4]
          elif row[3] == 'RBD':
            d1['eukaryotic(plants)']['RBD'][row[0]] = row[4]
          elif row[3] == 'terminator':
            d1['eukaryotic(plants)']['term'][row[0]] = row[4]

      elif row[2] == 'mammal':
        if row[3] == 'fullpro':
            d1['eukaryotic(mammals)']['prom']['fullpro'][row[0]] = row[4]
        elif row[3] == 'minpro':
          d1['eukaryotic(mammals)']['prom']['minpro'][row[0]] = row[4]
        elif row[3] == 'RBD':
          d1['eukaryotic(mammals)']['RBD'][row[0]] = row[4]
        elif row[3] == 'terminator':
          d1['eukaryotic(mammals)']['term'][row[0]] = row[4]

      elif row[3] == 'box':
        d1['box'][row[0]] = row[4]


d1
```

```python
import ipywidgets as widgets
from ipywidgets import HBox, VBox



# ------------------------SELECTION-----------------------------------------------------------
tab_fullW = widgets.Checkbox(
    value=False,
    description='Full Promoter',
    disabled=False
)

tab_minW = widgets.Checkbox(
    value=False,
    description='Min Promoter',
    disabled=False
)

tab_freeW = widgets.Checkbox(
    value=False,
    description='Free',
    disabled=False
)
select_promoter = HBox(children=[ tab_fullW,
                                 tab_minW,
                                 tab_freeW])

# Option values
TF_full = tab_fullW.value
TF_min = tab_minW.value
TF_free = tab_freeW.value

# -----------------------------------------------------------------------------------

# PROMOTER SECTION ####################################################################

# Constitutive Option ------------------------------------------------------
tab_promoter_constitutiveW = widgets.Dropdown(
    options= d1['prokaryotic']['prom']['fullpro'].keys(),
    value = None,
    description='Full Promoter:',
    disabled = not bool(TF_full)
)

def select_mode(tab_full ,tab_promoter_constitutive):
    tab_promoter_constitutiveW.disabled = not tab_full
    print('csscs',tab_fullW.value)

a = widgets.interactive(select_mode, 
                        tab_full = tab_fullW , 
                        tab_promoter_constitutive = tab_promoter_constitutiveW)


#   Inducible Option------------------------------------------------------
tab_promoter_inducible_boxW = widgets.Dropdown(
    options=d1['box'].keys(),
    value= None,
    description='Select Box:',
    disabled= not bool(TF_min))


tab_promoter_inducible_minpromoterW = widgets.Dropdown(
    options=d1['prokaryotic']['prom']['minpro'].keys(),
    value= None,
    #rows=10,
    description='Min promoter:',
        disabled= not bool(TF_min)
)




def select_mode(tab_min ,tab_promoter_inducible_box, tab_promoter_inducible_minpromoter):
    tab_promoter_inducible_boxW.disabled = not tab_min
    tab_promoter_inducible_minpromoterW.disabled = not tab_min
    print('inducible',tab_minW.value)

a = widgets.interactive(select_mode, 
                        tab_min = tab_minW , 
                        tab_promoter_inducible_box = tab_promoter_inducible_boxW,
                        tab_promoter_inducible_minpromoter = tab_promoter_inducible_minpromoterW)


#     To join previous optionsselect_promoter
tab_promoter_inducibleW = HBox(children=[ tab_promoter_inducible_boxW,
                                        tab_promoter_inducible_minpromoterW                                     
                                     ])
# -------------------------------------------------------------------------------

# ---2.3. Free Option
tab_promoter_freeW = widgets.Text(
    value='Promoter',
    placeholder='Type something',
    description='Write sequence:',
    disabled=False
)


tab2 = widgets.Accordion(children=[select_promoter,
                                   tab_promoter_constitutiveW,
                                    tab_promoter_inducibleW,
                                    tab_promoter_freeW])

tab2.set_title(0, 'Selection')
tab2.set_title(1, 'Full Promoter')
tab2.set_title(2, 'Min Promoter')
tab2.set_title(3, 'Free')


# 3. RBS---------------------------------------------------------------------------

tab_RBS_selection = widgets.SelectMultiple(
    options=['LecA', 'TerO'],
    value=['LecA'],
    #rows=10,
    description='Select Box:',
    disabled=False,
)


tab_RBS_free = widgets.Textarea(
    value='Hello World',
    placeholder='Sequence',
    description='String:',
    disabled=False
)
#     
tab_RBS = HBox(children=[tab_RBS_selection,
                                tab_RBS_free])


# 4. Gene selection

tab_gene = widgets.Textarea( # solo texto
    value='Hello World',
    placeholder='Sequence',
    description='String:',
    disabled=False
)

# 5. Terminator selection
tab_terminator_selection = widgets.SelectMultiple(
    options=['LecA', 'TerO'],
    value=['LecA'],
    #rows=10,
    description='Select Box:',
    disabled=False,
)


tab_terminator_free = widgets.Textarea(
    value='Hello World',
    placeholder='Sequence',
    description='String:',
    disabled=False
)
#     
tab_terminator = HBox(children=[tab_terminator_selection,
                                tab_terminator_free])


# TABS ASIGNMENT



tab3 = tab_RBS
tab4 = tab_gene
tab5 = tab_terminator

# Creation to visualizase
tab_nest = widgets.Tab()
tab_nest.children = [ tab2, tab3, tab4, tab5] # Agrupar Tabs
tab_nest.set_title(0, 'Promoter')
tab_nest.set_title(1, 'RBS')
tab_nest.set_title(2, 'Gene')
tab_nest.set_title(3, 'Terminator')
tab_nest # Mostrar Tab
```
